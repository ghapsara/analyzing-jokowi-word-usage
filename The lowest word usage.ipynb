Finding what happened at the low word usage hour

```python
import pandas as pd
```

```python
tweets_df = pd.read_json('data/normalized_tweets.json')
tweets_df['hour'] = tweets_df.datetime.apply(lambda d: d.hour)
tweets_df['word'] = tweets_df.message.apply(lambda m: m.split(' '))
tweets_df['word_length'] = tweets_df.word.apply(lambda w: len(w))
```

Low word usage at 2 AM

```python
tweets_at_2 = tweets_df[tweets_df.hour == 2]
tweets_at_2
```

Low word usage at 1 AM

```python
tweets_at_1 = tweets_df[tweets_df.hour == 1]
tweets_at_1
```

```python
tweets_at_3 = tweets_df[tweets_df.hour == 3]
tweets_at_3
```
