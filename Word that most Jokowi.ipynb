```python
import pandas as pd
```

#### reading data

```python
normalized_tweets = pd.read_json('data/normalized_tweets.json')
```

#### generate words
messages column => long message => split into words

```python
messages = list(normalized_tweets.message)
long_messages = ' '.join(messages)
words = long_messages.split(' ')
```

#### counting total word appearance

```python
words_df = pd.DataFrame(words, columns=['word'])
words_df = pd.DataFrame(words_df.word.value_counts())
words_df.reset_index(inplace=True)

words_df.columns = ['word', 'count']
```

```python
words_df
```

```python
## generating top 100 mentioned words
with open('results/words_most_jokowi.json', 'w', encoding='utf-8') as outfile:
   outfile.write(words_df[:100].to_json(orient='records'))
```

```python
# generate unique words
with open('results/tweet_words.json', 'w', encoding='utf-8') as outfile:
    outfile.write(words_df.word.to_json(orient='records'))
```
